```python
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pytz

# Set the time zone to Japan
japan_tz = pytz.timezone('Asia/Tokyo')

# Download the data for Japanese government bonds
jgb_data = yf.download('^JGB', start='2020-01-01', end='2021-12-31')

# Plot the interest rate movement
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(jgb_data.index, jgb_data['Close'])
ax.set_title('Japanese Government Bond Interest Rate')
ax.set_xlabel('Date')
ax.set_ylabel('Interest Rate')
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.xaxis.set_minor_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%d'))
ax.tick_params(axis='x', which='both', bottom=True, labelbottom=True, rotation=45)
ax.grid(True)
plt.show()
```
